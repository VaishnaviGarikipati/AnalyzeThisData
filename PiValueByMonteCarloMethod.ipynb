Write a Python program to simulate the calculation of π using Monte Carlo method. Your program draws the circle and the square, and plot the random points in the graph as shown in Figure 1. You need simulate different numbers of points (at least include 10, 500, 1,000, 5,000, 10,000, 50,000, 80,000, 100,000,
200,000) and calculate the value π for each different number of points. Then
draw a line chart to show the change of the accuracy of the value of π calculated
when the number of points is increased.
Submit Python code (2 points, no point will be given for a program that
doesn’t correctly implement the simulation algorithm), and simulation results
including the graph plots (1 point), π values (1 point), and line chart (1 point).

```python
import numpy as np
import matplotlib.pyplot as plt

def pi_cal(n):
    origin = (1,1)
    np.random.seed(0)
    x = np.random.uniform(low=origin[0]-1, high=origin[0]+1, size=[n,1])
    y = np.random.uniform(low=origin[1]-1, high=origin[1]+1, size=[n,1])
    inside_circle= (x-origin[0])**2+(y-origin[1])**2<1
    pi_value=4*np.sum(inside_circle)/n
    x_co_ordinates_inside_circle=x[inside_circle]
    y_co_ordinates_inside_circle=y[inside_circle]
    plt.figure(figsize=[5,5])
    plt.scatter(x,y,s=1)
    plt.scatter(x_co_ordinates_inside_circle,y_co_ordinates_inside_circle,color="r",s=1)
    plt.show()
    print("Pie value for "+str(n)+" points is: "+str(pi_value))
    return pi_value
   
N = [10,500,1000,5000,10000,50000,80000,100000,200000]


#Plots for each simulation
corresponding_pi = [pi_cal(i) for i in N]

#Line plot

#plt.set_xlim([0, 5])
plt.ylim([min(corresponding_pi)-0.1, max(corresponding_pi)+0.1])
plt.plot(N,corresponding_pi)
plt.show()
```
